```python
#creating a new data frame
df_2 = pd.DataFrame()

#Creating new variables 
#summing and multiplying
df_2['sum1']=(df_1.iloc[:, 0:999].sum(axis=1))*1000
df_2['sum2']=(df_1.iloc[:, 1000:1999].sum(axis=1))*2000
df_2['sum3']=(df_1.iloc[:, 2000:2999].sum(axis=1))*3000
df_2['sum4']=(df_1.iloc[:, 3000:3999].sum(axis=1))*4000
df_2['sum5']=(df_1.iloc[:, 4000:4999].sum(axis=1))*5000
df_2['sum6']=(df_1.iloc[:, 5000:5999].sum(axis=1))*6000
df_2['sum7']=(df_1.iloc[:, 6000:6999].sum(axis=1))*7000
df_2['sum8']=(df_1.iloc[:, 7000:7999].sum(axis=1))*8000
df_2['sum9']=(df_1.iloc[:, 8000:8999].sum(axis=1))*9000
df_2['sum10']=(df_1.iloc[:, 9000:9999].sum(axis=1))*10000
```

```python
#Standardizing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df_2)
x = pd.DataFrame(scaler.transform(df_2))
```

```python
#Implementing the K_means

from sklearn.cluster import KMeans  #importing 

kmeans = KMeans(n_clusters = 8)
#Implemeanting K-Means 
k_fit = kmeans.fit(x)
#labeling the clusters
k_s  =  kmeans.predict(x)
#centers
centers = kmeans.cluster_centers_
```

```python
df_2['k_clust'] = k_s
```
